```python
import numpy as np
import pandas as pd
```

```python
a = pd.read_csv('oscar_nominees_with_fixed_race.csv')
a['category'].replace(["ACTRESS IN A LEADING ROLE"], "Actress In A Leading Role", inplace=True)
a['category'].replace(["ACTOR IN A LEADING ROLE"], "Actor In A Leading Role", inplace=True)
a['category'].replace(["ACTRESS IN A SUPPORTING ROLE"], "Supporting Actress In A Leading Role", inplace=True)
a['category'].replace(["ACTOR IN A SUPPORTING ROLE"], "Supporting Actor In A Leading Role", inplace=True)

a.head()
```

```python
b = pd.read_csv('golden_globe_merged.csv')
b.head()
```

```python
c = pd.read_csv('sag_merged.csv')
c.head()
```

```python
acount = a.groupby(['year_ceremony', 'fixed_race', 'category']).count().race.rename('count').reset_index()
acount['award'] = 'oscars'
acount.head()
```

```python
b.groupby(['year_award', 'race', 'category']).count()
```

```python
b['fixed_race'] =  b.race.replace({"White [1]": 'white', 
                'Multiracial':'other',
                'Multiracial [1]':'other',
                'Multiracial [2]': 'other',
                'Asian/Indian': 'asian',
                'American Aborigine': 'other',
                'American Aborigine [1]': 'other',
                'e': 'other',
                'Middle Eastern':'White'}).str.lower()
```

```python
b['category'] = b.apply(lambda row: row['category'].strip(), axis=1)
b['category'].replace(["Best Performance by an Actress in a Motion Picture - Drama", "Best Performance by an Actress in a Motion Picture - Musical or Comedy"], "Actress In A Leading Role", inplace=True)
b['category'].replace(["Best Performance by an Actor in a Motion Picture - Drama", "Best Performance by an Actor in a Motion Picture - Musical or Comedy"], "Actor In A Leading Role", inplace=True)
b['category'].replace(["Best Performance by an Actress in a Supporting Role in any Motion Picture"], "Supporting Actress In A Leading Role", inplace=True)
b['category'].replace(["Best Performance by an Actor in a Supporting Role in any Motion Picture"], "Supporting Actor In A Leading Role", inplace=True)

bcount = b.dropna().groupby(['year_award', 'fixed_race', 'category']).count().race.rename('count').reset_index()
bcount['award'] = 'golden globes'
bcount.columns = ['year_ceremony'] + list(bcount.columns[1:])
categories = ["Actress In A Leading Role", "Actor In A Leading Role", "Supporting Actress In A Leading Role", 
             "Supporting Actor In A Leading Role"]
bcount = bcount[bcount['category'].isin(categories)]
bcount['category'].value_counts()
```

```python
c['fixed_race'] = c.race.replace({"Middle Eastern": 'White',
               'Asian/Indian': 'Asian',
               'Multiracial': 'Other',
               'Multiracial [1]': 'Other',
               'Multiracial [2]': 'Other'}).str.lower()
```

```python
c['year_ceremony'] = c.year_ceremony.apply(lambda x: x.split('-')[0]).astype(int)
```

```python
c['category'] = c.apply(lambda row: row['category'].strip(), axis=1)
c['category'].replace(['FEMALE LEAD ROLE', 'FEMALE LEAD', 'FEMALE LEAD IN A MOTION PICTURE', 'FEMALE ACTOR IN A LEADING ROLE'], 'Actress In A Leading Role', inplace=True)
c['category'].replace(['MALE LEAD ROLE', 'MALE LEAD', 'MALE LEAD IN A MOTION PICTURE', 'MALE ACTOR IN A LEADING ROLE'],'Actor In A Leading Role' , inplace=True)
c['category'].replace(['FEMALE SUPPORTING ROLE', 'FEMALE SUPPORT', 'FEMALE SUPPORT IN A MOTION PICTURE',  'FEMALE ACTOR IN A SUPPORTING ROLE'], "Supporting Actress In A Leading Role", inplace=True)
c['category'].replace(['MALE SUPPORTING ROLE', 'MALE SUPPORT', 'MALE SUPPORT IN A MOTION PICTURE', 'MALE ACTOR IN A SUPPORTING ROLE'], "Supporting Actor In A Leading Role", inplace=True)

ccount = c.groupby(['year_ceremony','fixed_race', 'category']).count().race.rename('count').reset_index()
ccount['award'] = 'sag'
categories = ["Actress In A Leading Role", "Actor In A Leading Role", "Supporting Actress In A Leading Role", 
             "Supporting Actor In A Leading Role"]
ccount = ccount[ccount['category'].isin(categories)]
```

```python
df = pd.concat([acount, bcount, ccount])
df.fixed_race.unique()
```

```python
df.to_csv('all_awards_race_year_count.csv', index=False)
```

```python
df.groupby(['year_ceremony','fixed_race', 'award', 'category'])['count'].apply(lambda x: x.astype(int)/x.astype(int).sum())
```

```python
# df.groupby(['year_ceremony','fixed_race', 'award'])['count'].apply(lambda x: x.astype(int)/x.astype(int).sum())
df.groupby(['year_ceremony','award', 'category'])['count'].apply(lambda x: x.astype(int)/x.astype(int).sum())
```

```python
df['pct_race'] = df['count'] / df.groupby(['year_ceremony','award', 'category'])['count'].transform('sum')
```
