```python
import numpy as np
import pandas as pd
```

```python
a = pd.read_csv('oscar_nominees_with_fixed_race.csv')
a.head()
```

```python
b = pd.read_csv('golden_globe_merged.csv')
b.head()
```

```python
c = pd.read_csv('sag_merged.csv')
c.head()
```

```python
acount = a.groupby(['year_ceremony', 'fixed_race']).count().race.rename('count').reset_index()
acount['award'] = 'oscars'
acount.head()
```

```python
b.groupby(['year_award', 'race']).count()
```

```python
b['fixed_race'] =  b.race.replace({"White [1]": 'white', 
                'Multiracial':'other',
                'Multiracial [1]':'other',
                'Multiracial [2]': 'other',
                'Asian/Indian': 'asian',
                'American Aborigine': 'other',
                'American Aborigine [1]': 'other',
                'e': 'other',
                'Middle Eastern':'asian'}).str.lower()
```

```python
bcount = b.dropna().groupby(['year_award', 'fixed_race']).count().race.rename('count').reset_index()
bcount['award'] = 'golden globes'
bcount.columns = ['year_ceremony'] + list(bcount.columns[1:])
bcount.head()
```

```python
c['fixed_race'] = c.race.replace({"Middle Eastern": 'Asian',
               'Asian/Indian': 'Asian',
               'Multiracial': 'Other',
               'Multiracial [1]': 'Other',
               'Multiracial [2]': 'Other'}).str.lower()
```

```python
c['year_ceremony'] = c.year_ceremony.apply(lambda x: x.split('-')[0]).astype(int)
```

```python
ccount = c.groupby(['year_ceremony','fixed_race']).count().race.rename('count').reset_index()
ccount['award'] = 'sag'
ccount.head()
```

```python
df = pd.concat([acount, bcount, ccount])
df.fixed_race.unique()
```

```python
df.to_csv('all_awards_race_year_count.csv', index=False)
```

```python
df.groupby(['year_ceremony','fixed_race', 'award'])['count'].apply(lambda x: x.astype(int)/x.astype(int).sum())
```

```python
# df.groupby(['year_ceremony','fixed_race', 'award'])['count'].apply(lambda x: x.astype(int)/x.astype(int).sum())
df.groupby(['year_ceremony','award'])['count'].apply(lambda x: x.astype(int)/x.astype(int).sum())
```

```python
df['pct_race'] = df['count'] / df.groupby(['year_ceremony','award'])['count'].transform('sum')
```
