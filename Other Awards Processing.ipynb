## Golden Globes

```python
import numpy as np
import pandas as pd
```

```python
go = pd.read_csv("golden_globe_awards.csv")
os = pd.read_csv("oscar_nominees_with_fixed_race.csv", index_col=0)
os.head()
```

```python
go.head()
```

```python
os = os[os['year_film']>1943]
```

```python
os_actors = set(os.name.unique())
go_actors = set(go.nominee.unique())
len(go_actors - os_actors)
```

```python
# 806 actors in Golden Globes covered by Oscars
print(len(go_actors), len(os_actors), len(go_actors - os_actors), len(go_actors.intersection(os_actors)))
```

```python
go.columns = ['year_film', 'year_ceremony', 'ceremony','category','name','film','winner']
go.head()
```

```python
go = go.merge(os[['name','race','fixed_race']], how='left').drop_duplicates()
go.head()
```

```python
go[['race','fixed_race']] = go[['race','fixed_race']].fillna(value='unknown')
```

```python
go.head()
```

```python
categories = ['Best Director - Motion Picture', 'Best Performance by an Actor in a Motion Picture - Drama',
              'Best Performance by an Actress in a Motion Picture - Drama',
              'Best Performance by an Actress in a Supporting Role in a Series, Limited Series or Motion Picture Made for Television',
       'Best Performance by an Actor in a Supporting Role in a Series, Limited Series or Motion Picture Made for Television',]
```

```python
go_temp = go[go.category.isin(categories)]
go_temp.shape
```

```python
go_temp.to_csv("golden_globe_awards_with_race_temp.csv", index=False)
```
